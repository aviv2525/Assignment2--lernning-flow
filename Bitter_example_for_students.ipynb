```python
threshold = len(df) * 0.5
df = df.drop(columns=[col for col in df.columns if (df[col] == 0).sum() > threshold])
```

```python
df.shape
```

```python
import numpy as np

# חישוב מטריצת המתאם
corr_matrix = df.corr().abs()

# יצירת מסכה לחלק העליון של המטריצה
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# זיהוי עמודות עם מתאם גבוה (למשל, מעל 0.9)
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

# הסרת עמודות עם מתאם גבוה
df.drop(columns=to_drop, inplace=True)
```
